```python
with open('../Claves/aemet.key') as archivo:
    api_key = archivo.readline().rstrip('\n')
```

```python
import json
import requests as req
#from aemet import Aemet
```

```python
# URL navegador

'''CAMBIAR FECHAS INICIO Y FECHA FIN

    DOS DIAS ATRAS NO SUELE MOSTRAR DATOS'''

fecha_inicio = '2022-08-01'
fecha_fin = '2022-08-31'

url_navegador = f'https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/{fecha_inicio}T00:00:00UTC/fechafin/{fecha_fin}T23:59:59UTC/todasestaciones/?api_key={api_key}'
```

```python
res_navegador = req.get(url_navegador)
res_navegador
```

```python
url = res_navegador.json()['datos']
url
```

```python
res = req.get(url)
res
```

```python
tiempo = res.json()
tiempo
```

```python
for e in range(len(tiempo)):
    colec.insert_one(tiempo[e])
```
